## Scipy Signal get_window(...)

```python
from scipy import signal
import numpy as np
```

```python
#@title
%%html
<iframe src="https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.get_window.html" height="500" width="900" title="Iframe Example"></iframe>
```

### Create a Window

```python
window = signal.get_window(('general_gaussian', 1, 5),32)
```

### Plot the Window

```python
import plotly.express as px
n=np.arange(window.size)
fig = px.line(x=n, y=window, labels={'x':'[n]', 'y':'Amplitude'})
fig.show()
```

### Calculate the Frequency Response

```python
w, h = signal.freqz(window, worN=1024)
h/=np.abs(h).max()
angles = np.unwrap(np.angle(h))
```

### Plot the Frequency Response

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
fig = make_subplots(rows=1, cols=2,  subplot_titles=("Magnitude (normalized)", "Phase"))

fig.add_trace(
    go.Scatter(x=w/np.pi, y=20 * np.log10(abs(h)), mode='lines'),
    row=1, col=1, 
)

fig.add_trace(
    go.Scatter(x=w/np.pi, y=angles,mode='lines'),
    row=1, col=2
)
fig['layout']['xaxis']['title']='$\omega \cdot \pi$'
fig['layout']['xaxis2']['title']='$\omega \cdot \pi$'
fig['layout']['yaxis']['title']='Magnitude [dB]'
fig['layout']['yaxis2']['title']='Angle [radians]'
fig.update_layout(height=600, width=800, title_text="Frequency Response")
fig.update(layout_showlegend=False)

fig.show()
```
